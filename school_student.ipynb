```python
import pandas as pd
from pathlib import Path
```

```python
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
school_amount = school_data_complete["school_name"].unique()
len(school_amount)
#result shows there are 15 different schools
```

```python
#result shows there are 39170 different students 
student_amount = len(student_data["student_name"])
student_amount
```

```python
total_budget = school_data["budget"].sum()
total_budget
#result shows the total budget is 24,649,428
```

```python
avg_math = student_data["maths_score"].mean()
avg_math
#result shows the average maths score is 70.33819249425581
```

```python
avg_reading = student_data["reading_score"].mean()
avg_reading
#result shows the average reading score is 69.98013786060761
```

```python
pass_maths = student_data.loc[student_data["maths_score"] >= 50]
math_passers=len(pass_maths["maths_score"])

print(math_passers/student_amount *100)
#result shows 86.07863160582077 percent of students passed maths
```

```python
pass_reading = student_data.loc[student_data["reading_score"] >= 50]
reading_passers = len(pass_reading["reading_score"])

print(reading_passers/student_amount * 100)
#result shows 84.42685728874139 percent of students passed reading
```

```python
new = student_data.loc[(student_data["reading_score"] >= 50) & (student_data["maths_score"] >= 50), :]
new
len(new)
len(new)/len(student_data) * 100

#result shows 72.80827163645647 percent of students passed both reading and maths
```
